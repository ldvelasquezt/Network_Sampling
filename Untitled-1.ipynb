Sampling procedures can get us samples while estimation framework is also necessary if we need to estimate specific properties of the original graph. For estimation framework, as the stationary distributions are known or able to be calculated when the random walks are designed, the probability of being chosen is known, so is the weight of the sample.
A commonly used principle to construct unbiased estimators is employing Horvitz Thompson (HT) estimator (Horvitz and Thompson, 1952) for the function of interest. HT estimator is widely used in many papers (Lee, Xu and Eun, 2012; Li et al., 2015, 2019; Zhou, Zhang and Das, 2015; Zhao et al., 2019).

```python
def montecarlo_estimacion_method1(n,k):
    try:
        # Check if the argument is an integer
        if not isinstance(n, int):
            raise TypeError("Argument must be an integer")
        

        # Process
          # make an object and call function MHRW
       
        # Initialize an empty list to store the results
        results = []
        timeout = time.time() + 5
        # Repeat the process n times and store the result in the list
        for i in range(n): 
             
            if time.time() > timeout:
                 break
            
            object1=Graph_Sampling.SRW_RWF_ISRW()
            
            sample1 = object1.random_walk_sampling_simple(g,k) # graph, number of nodes to sample
            
            samplindasg=list(sample1)
            df_sampled=df[df['Label'].isin(samplindasg)]  
            grouped_df = df_sampled.groupby('CLUSTER').agg({'Piricularia': 'mean', 'Label': 'count'}) 
            grouped_df['total'] = grouped_df['Piricularia'] * grouped_df['Label'] 
            total_sample=grouped_df['Label'].sum()
            estimation= grouped_df['total'].sum()/230
            result = estimation# process to be repeated
            results.append(result)
           
        
        # Return the list of results
        estimation=np.mean(result)
        sesgo=(estimation-0.0198486167565217)**2
       
        varianza=((estimation-0.0198486167565217)**2)/total_sample
        
        rmse=varianza+sesgo**2
        cv= np.sqrt(varianza)/estimation
        print(estimation)
        print(sesgo)
        print(varianza)
        print(rmse)
        print(cv*100)

        return varianza,sesgo
    except TypeError as e:
        # Log the error
        print(f"Error: {e}")
        return []
```
